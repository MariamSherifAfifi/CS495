```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

#open the url 
#read the content
#store in the beautifulSoup object
url = "https://careerfoundry.com/en/blog/ux-design/how-to-become-a-ux-designer/"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
#return any content. exmaple:images
soup.find_all("img")
```

```python
#retrun any content. example:links.
soup.find_all("a")

```

```python
#count the numbers of links in the web page
sum = 0
for tag in soup.find_all("a"):
    sum=sum+1
    
```

```python
#print the totall number of the links in this web page
print('totall number of links:' ,sum)
```

```python
#print all text (words)
print(soup.get_text())
```

```python
#stores number of words
word = 0
for tag in soup.get_text():
    word=word+1
```

```python
print('total number of words:' , word);
```
